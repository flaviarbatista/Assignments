### **TODO: Check if the book is in the list and return True if available, False otherwise**

```python
def check_availability(book_name):
    """
    Return True if book_name is present in the global books list, False otherwise.
    Comparison is case-insensitive and trims whitespace.
    """
    if not book_name:
        return False
    target = book_name.strip().lower()
    for b in books:
        if b and b.strip().lower() == target:
            return True
    return False

# Test the function
print(f"Is 'Python Basics' available? {check_availability('Python Basics')}")
```

##  6. File I/O

```python
def save_books_to_file(filename):
    with open(filename, 'w') as f:
        for book in books:
            f.write(f"{book}\n")
```

### **TODO: Implement a function to read books from a file**

```python
def read_books_from_file(filename):
    """
    Read book titles from filename and return a list of titles.
    Ignores blank lines. Returns empty list if file missing or unreadable.
    """
    try:
        with open(filename, 'r') as f:
            books_read = [line.strip() for line in f if line.strip()]
        print(f"Read {len(books_read)} books from '{filename}'.")
        return books_read
    except FileNotFoundError:
        print(f"File not found: {filename}")
        return []
    except Exception as e:
        print("Error reading file:", e)
        return []
```

## 7. Pickle for serialization

```python
def save_books_pickle(filename):
    with open(filename, 'wb') as f:
        pickle.dump(books, f)
```

### **TODO: Implement a function to load books using pickle**

```python
def load_books_pickle(filename):
    """
    Load and return the list of books from a pickle file.
    Returns an empty list if the file is missing or an error occurs.
    """
    try:
        with open(filename, 'rb') as f:
            data = pickle.load(f)
        print(f"Loaded {len(data)} books from '{filename}'.")
        return data
    except FileNotFoundError:
        print(f"Pickle file not found: {filename}")
        return []
    except Exception as e:
        print("Error loading pickle:", e)
        return []
```

### Test pickle functions

```python
save_books_pickle("books.pkl")
print("Books loaded from pickle:", load_books_pickle("books.pkl"))
```

## 8. SQLite Database Interaction

```python
def create_books_table():
    conn = sqlite3.connect('library.db')
    cursor = conn.cursor()
    cursor.execute('''CREATE TABLE IF NOT EXISTS books
                      (id INTEGER PRIMARY KEY, title TEXT, quantity INTEGER)''')
    conn.commit()
    conn.close()
```

### **TODO: Implement a function to add a book to the SQLite database**

```python
def add_book_to_db(title, quantity):
    """
    Add a book to the 'books' table. Simple implementation: insert a new row.
    """
    if not title:
        print("Empty title - not adding.")
        return

    try:
        qty = int(quantity)
    except Exception:
        print("Invalid quantity - must be an integer.")
        return

    try:
        conn = sqlite3.connect('library.db')
        cur = conn.cursor()
        cur.execute("INSERT INTO books (title, quantity) VALUES (?, ?)", (title.strip(), qty))
        conn.commit()
        print(f"Inserted '{title.strip()}' with quantity {qty}.")
    except Exception as e:
        print("Database error:", e)
    finally:
        conn.close()
```

### Test database functions

## 9. Web API Interaction (GitHub API)

```python
def get_python_repos():
    url = "https://api.github.com/search/repositories"
    params = {"q": "language:python", "sort": "stars", "order": "desc"}
    response = requests.get(url, params=params)
    data = response.json()
    return [repo['name'] for repo in data['items'][:5]]
```

### Test API function

```python
print("Top 5 Python repositories on GitHub:", get_python_repos())

print("\nLibrary Management System operations completed.")
```
